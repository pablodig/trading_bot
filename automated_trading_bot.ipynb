```python

import alpaca_trade_api as tradeapi
from datetime import datetime
import os

# Setup Alpaca API
API_KEY = 'your_api_key'
SECRET_KEY = 'your_secret_key'
BASE_URL = 'https://paper-api.alpaca.markets'

api = tradeapi.REST(API_KEY, SECRET_KEY, base_url=BASE_URL, api_version='v2')

def get_portfolio_stock_quantity(symbol):
    try:
        position = api.get_position(symbol)
        return int(position.qty)
    except tradeapi.rest.APIError as e:
        print(f"No existing position for {symbol} found.")
        return 0

def execute_trade(action, symbol, quantity):
    if action == 'buy':
        api.submit_order(symbol=symbol, qty=quantity, side='buy', type='market', time_in_force='gtc')
    elif action == 'sell':
        owned_quantity = get_portfolio_stock_quantity(symbol)
        if owned_quantity >= quantity:
            api.submit_order(symbol=symbol, qty=quantity, side='sell', type='market', time_in_force='gtc')
        else:
            print(f"Attempted to sell {quantity} shares of {symbol}, but only own {owned_quantity} shares.")

def daily_trade():
    symbol = 'AAPL' # Example stock symbol
    action = 'buy' # or 'sell', determined by your strategy
    quantity = 1 # Determine based on strategy and portfolio checks

    execute_trade(action, symbol, quantity)

if __name__ == '__main__':
    daily_trade()
```
